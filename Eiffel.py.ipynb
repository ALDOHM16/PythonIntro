```python
Monedas y la Torre Eiffel
Un día encuentras una moneda de un euro en la acera junto a la torre Eiffel en Paris. 
Si cada día a partir de entonces, vas y colocas una moneda adicional (una sobre la otra). 
¿Cuánto tiempo tardas para que la columna de monedas en llegue a ser igual a la altura de la torre?
```

```python
# eiffel.py
moneda_grosor = 0.002125 # Metros (2.125 mm)
eiffel_altura = 300 # Altura de la torre (metros)
num_monedas = 10
dia = 1

while num_monedas * moneda_grosor < eiffel_altura:
    print(dia, num_monedas, num_monedas * moneda_grosor)
    dia = dia + 1
    num_monedas = num_monedas + 10

print('Número de días', dia)
print('Número de monedas', num_monedas)
print('Altura final', num_monedas * moneda_grosor)
print('Número de años', dia/365)
```
